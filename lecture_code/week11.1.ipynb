```python
import pandas as pd

df = pd.read_csv("/workspaces/sp26_data_3500/lecture_code/titanic.csv")
df.head()
```

```python
# data cleaning
df.info()
```

```python
# drop missing values
df = df.drop(columns=["Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].median())
df = df.dropna(subset=["Embarked"])

df.info()
```

```python
# rename columns
df = df.rename(columns={"Pclass":"Passenger_Class"})
df = df.rename(columns={"SibSp":"Siblings/Spouses"})
df = df.rename(columns={"Parch":"Parents/Children"})
df = df.rename(columns={"PassengerId":"Passenger_ID"})
df.head()
```

```python
"""
Programming Activity 3

Using the Titanic dataset and the Pandas library, find the average survival rate by class and sex.
"""

survival_rate = df.groupby(["Passenger_Class", "Sex"])["Survived"].mean()
print("Avg survival rate by class and sex", survival_rate)
```
